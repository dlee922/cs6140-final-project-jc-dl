```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import sys
sys.path.append('../../')  # go up two levels to root
from sklearn.multioutput import MultiOutputClassifier, ClassifierChain
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd
import numpy as np
from utils.utils import nested_cv, load_train_test
```

```python
X_train, X_test, y_train, y_test = load_train_test()
```

```python
# use uniform priors to fix imbalance
order = [2,1,3,0,5,6,4]
clf = ClassifierChain(LinearDiscriminantAnalysis(shrinkage='auto', solver='lsqr', priors=[0.5,0.5]), order=order)
clf.fit(X_train, y_train) # fit data on training fold
y_pred = clf.predict(X_test) # test 
print(f1_score(y_test, y_pred, average='macro', zero_division=0))
print(accuracy_score(y_test, y_pred))
```
