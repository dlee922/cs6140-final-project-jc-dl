```python
import pandas as pd

# Load the data
clinical = pd.read_csv('LUAD_clinicalMatrix', sep='\t')
survival = pd.read_csv('LUAD_survival.txt', sep='\t')

# print(clinical.shape)
# print(clinical.head())
# print(clinical.columns.tolist())

# print(clinical['gender'].value_counts())
# print(clinical['tobacco_smoking_history'].describe())
survival = survival.rename(columns={'sample': 'sampleID'})

df = pd.merge(clinical, survival, on='sampleID', how='inner')
# print(df)
# print(df.shape)
# print(df.columns.to_list())

# check outcome variable
print(df['OS'].value_counts())       # overall survival status (0/1)
print(df['OS.time'].describe())      # survival time in days

# check missing data
# print(df.isnull().sum().sort_values(ascending=False).head(20))

print(df['gender'].value_counts())
print(df['age_at_initial_pathologic_diagnosis'].describe())
print(df['OS'].value_counts())
```
